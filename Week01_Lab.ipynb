### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# Your code here
#Total episodes 
episodes_per_season=10

#Time for season 1 
season1=(45*60+35)*episodes_per_season

#Time for season2-5
season2_5=(35*60+15)*episodes_per_season*4

#Total time in seconds 
total_seconds=season1+season2_5

#Converting to hours minutes and seconds 
hours=total_seconds//3600 #1 hour has 3600 seconds 
minutes=(total_seconds % 3600) // 60 
seconds=total_seconds % 60 

print("Total Time:", hours, "h", minutes, "m", seconds, "s")
````

### Excellent 2. Data Science: Mini Statistics Report

Ask for five numeric values (space-separated). Print the values, total, mean, median, and sample standard deviation.

**Example Output:**
```
Values: 80, 90, 70, 85, 75
Total = 400.00
Mean = 80.00
Median = 80.00
Std Dev = 7.91
```


*Optional hint:* `import statistics as stats`


````python
# Your code here
import statistics as stats 
nums=(input("Enter five numbers separated by spaces:"))
values=nums.split()#cuts the text whereever there is a space 
values=[float(x) for x in values]#float(x)converts the text into numbers with decimal
total=sum(values)
mean=stats.mean(values)
median=stats.median(values)
std_dev=stats.stdev(values)
print("Values:", values)
print("Total:", total)
print("Mean:", mean)
print("Median:", median)
print(f"Std dev={std_dev:.2f}")
````

### Excellent 3. IT: Path Decomposition Report

Ask for a file or directory path and print its absolute path, parent, name, stem, and suffix.

**Example Output:**
```
Absolute path: /home/user/project/service.log
Parent: /home/user/project
Name: service.log
Stem: service
Suffix: .log
```


*Optional hint:* `from pathlib import Path`


````python
# Your code here
from pathlib import Path 
user=input("Enter a file or directory path:")
p=Path(user)#wraps that string in a path object 
print("Absolute path: ", p.resolve())#gives absolute fully resolved path
print("Parent:", p.parent)#gives the folder 
print("Name:", p.name)#last part of the path(file or folder)
print("Stem:", p.stem)#gives the name without extension
print("Suffix:",p.suffix)#gives the extension

````
